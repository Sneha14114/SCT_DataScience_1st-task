```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("/content/RTA Dataset.csv")

# Basic cleanup
df['Time'] = pd.to_datetime(df['Time'], errors='coerce')
df['Hour'] = df['Time'].dt.hour
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Hour', data=df, palette='rocket')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Day_of_week', data=df, order=[
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='mako')
plt.title('Accidents by Day of Week')
plt.xlabel('Day')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Accident_severity', data=df, palette='Set2')
plt.title('Distribution of Accident Severity')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(x='Driving_experience', hue='Accident_severity', data=df, palette='coolwarm')
plt.title('Driving Experience vs. Accident Severity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 5))
sns.countplot(x='Type_of_vehicle', hue='Accident_severity', data=df, palette='pastel')
plt.title('Vehicle Type vs. Accident Severity')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
top_causes = df['Cause_of_accident'].value_counts().nlargest(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_causes.index, y=top_causes.values, palette='flare')
plt.title('Top 10 Causes of Accidents')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_causes = df['Cause_of_accident'].value_counts().nlargest(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_causes.index, y=top_causes.values, palette='flare')
plt.title('Top 10 Causes of Accidents')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Convert selected categorical features to dummy vars for correlation
categorical_features = ['Accident_severity', 'Day_of_week', 'Sex_of_driver']
df_corr = pd.get_dummies(df[categorical_features], drop_first=True)

plt.figure(figsize=(10, 6))
sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Selected Categorical Features")
plt.tight_layout()
plt.show()
```
